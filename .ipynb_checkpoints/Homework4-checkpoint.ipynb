### Question 2.68
$$x_{\delta}(t) = \sum_{m=-\infty}^{\infty}(x(mT_s)\delta(t-mT_s) \rightarrow \boxed{h(t) = \Pi[(t-\frac{1}{2}T_s)/T_s]} \rightarrow y(t)$$
#### Figure 2.41

Figure 2.41 illustrates so-called zero-order-hold reconstruction.

(a) Sketch ùë¶(ùë°) for a typical ùë•(ùë°). Under what conditions is ùë¶(ùë°) a good approximation to ùë•(ùë°)
Conditions for good approximation y(t) to x(t):
- high sampling rate
- Nyquist is met: x(t) has no frequencies higher than f_max, and f_s >= 2f_max.
- x(t) is smooth. The smoother x(t) the better interpolation between samples.

See plot below for sketch.

(b) Find the spectrum ofùë¶(ùë°)in terms of the spectrum of ùë•(ùë°). Discuss the approximation of ùë¶(ùë°) to ùë•(ùë°) in terms of frequency-domain arguments.
$$x_{\delta}(t) = x(t)\sum[\delta(t - mT_s)]$$
$$X_{\delta}(\omega) = \frac{1}{2\pi}[X(\omega)\frac{2\pi}{T_s}\sum[\delta(\omega - n\omega_s]]$$
$$\omega_s = \frac{2\pi}{T_s}$$
$$X_{\delta}(\omega) = \frac{1}{T_s}[X(\omega)\sum[\delta(\omega - n\omega_s)]]$$

$$y(t) = x_{\delta}(t)*h(t)$$
$$Y(\omega) = X_{\delta}(\omega)H(\omega)$$
Frequency response of zero-order hold:
$$H(\omega) = T_ssinc(\omega T_s/2)e^{-j\omega T_s/2}$$
$$Y(\omega) = \frac{1}{T_s}[X(\omega)\sum[\delta(\omega - n\omega_s)]]T_ssinc(\omega T_s/2)e^{-j\omega T_s/2}$$

```python
def zero_order_hold(x, Ts, t):
    y = np.zeros_like(t)
    for i in range(len(t)):
        m = int(t[i] / Ts)
        if m*Ts >= 0 and m*Ts <= t[-1]:
          y[i] = x[int(m * Ts / (t[1]-t[0]))]
    return y

def x(t):
    return np.cos(t)

Ts = 0.2
t = np.linspace(0, 10, 512)
x1 = x(t)
y = zero_order_hold(x1, Ts, t)

plt.figure(figsize=(10, 6))
plt.plot(t, x1, label="x(t) (Original)")
plt.plot(t, y, label="y(t) (Zero-Order Hold)")
plt.title("Zero-Order Hold Reconstruction")
plt.xlabel("t")
plt.ylabel("x(t), y(t)")
plt.legend()
plt.grid()
plt.show()

fs = 1/Ts
f = np.linspace(-5, 5, 1024)
omega = 2*np.pi*f

H_omega = Ts * np.sinc(omega * Ts / 2) * np.exp(-1j * omega * Ts / 2)

N = 512
t_sampled = np.linspace(0, 5, N)
x_sampled = x(t_sampled)

X_omega = np.fft.fft(x_sampled)
X_omega = np.fft.fftshift(X_omega)
f_sampled = np.fft.fftfreq(N, d=t_sampled[1]-t_sampled[0])
f_sampled = np.fft.fftshift(f_sampled)

plt.figure(figsize=(10, 6))

plt.subplot(211)
plt.plot(f_sampled, np.abs(X_omega), label = "|X(œâ)|")
plt.title("Spectrum of Sampled x(t)")
plt.xlabel("f (Hz)")
plt.ylabel("|X(œâ)|")
plt.grid()
plt.legend()

plt.subplot(212)
plt.plot(f, np.abs(H_omega), label="|H(œâ)|")
plt.title("Spectrum of Zero-Order Hold")
plt.xlabel("f (Hz)")
plt.ylabel("|H(œâ)|")
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

### Question 2.71

$$y(t) = x(t)cos(\omega_0t) + \hat{x}sin(\omega_0t)$$
in terms of the spectrum X(f) of x(t), where X(f) is lowpass with bandwidth
$$B < f_0 = \frac{\omega_0}{2\pi}$$

Plot Y(f) for a typical X(f)

$$y(t) = x(t)\frac{1}{2}[e^{j\omega_0t}+e^{-j\omega_0t}]+\hat{x}(t)\frac{1}{2j}[e^{j\omega_0t}-e^{-j\omega_0t}]$$
$$Y(f) = \frac{1}{2}[X(f - f_0) + X(f + f_0)] + \frac{1}{2j}[\hat{X}(f - f_0) - \hat{X}(f + f_0)]$$

$$Y(f) = X(f - f_0) for f > 0$$
$$Y(f) = X(f + f_0) for f < 0$$

```python
def X(f):
    if abs(f) < B:
        return 1
    else:
        return 0

B = 5
## Plots will overlap if f0 <= B
f0 = 10
f = np.linspace(-20, 20, 1000)
Y = np.array([X(f_val - f0) if f_val > 0 else X(f_val + f0) for f_val in f])

plt.figure(figsize=(10, 6))
plt.plot(f, Y)
plt.title("Spectrum Y(f)")
plt.xlabel("Frequency (f)")
plt.ylabel("Y(f)")
plt.grid()
plt.show()
```
