## Question 2.8
Write the signal ùë•(ùë°) = cos(6ùúãùë°) + 2 sin(10ùúãùë°) as
- (a) The real part of a sum of rotating phasors.
$$x(t) = Re(1 * e^{j(6pi*t)}) + Re(2 * e^{j(10pi*t - pi/2)})$$
- (b) A sum of rotating phasors plus their complex conjugates.
$$x(t) = 1/2 * e^{j(6pi*t)} + 1/2 * e^{-j(6pi*t)} + 1 * e^{j(10pi*t - pi/2)} + 1 * e^{-j(10pi*t - pi/2)}$$

- (c) From your results in parts (a) and (b), sketch the single-sided and double-sided amplitude and phase spectra of ùë•(ùë°).


```python
import numpy as np
import matplotlib.pyplot as plt

# (c) From your results in parts (a) and (b), sketch the single-sided and double-sided amplitude and phase spectra of ùë•(ùë°).

f1 = 3  # 6pi / 2pi
f2 = 5  # 10pi / 2pi

A1 = 1/2  # For 6pi*t component
A2 = 1    # For 10pi*t component

# single sided
plt.figure(figsize=(8, 6))

# Amplitude
plt.subplot(2, 1, 1)
plt.stem([f1, f2], [A1, A2]) #stem plot for discrete values
plt.title("Single-Sided Amplitude Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)

# Phase
plt.subplot(2, 1, 2)
plt.stem([f1, f2], [0, -np.pi/2])
plt.title("Single-Sided Phase Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Phase (radians)")
plt.grid(True)

plt.tight_layout()
plt.show()

# double sided
plt.figure(figsize=(8, 6))

# Amplitude
plt.subplot(2, 1, 1)
plt.stem([-f2, -f1, f1, f2], [A2, A1, A1, A2])
plt.title("Double-Sided Amplitude Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)

# Phase
plt.subplot(2, 1, 2)
plt.stem([-f2, -f1, f1, f2], [np.pi/2, 0, 0, -np.pi/2])
plt.title("Double-Sided Phase Spectrum")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Phase (radians)")
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Question 2.9abc
Find the normalized power for each signal below that is a power signal and the normalized energy for each
signal that is an energy signal. If a signal is neither a power signal nor an energy signal, so designate it. Sketch
each signal (ùõº is a positive constant).

- (a) $$x1(t) = 2cos(4pi*t + 2*pi/3)$$
This signal is a periodic power signal.
$$P = A^2/2 = 2$$
- (b) $$x2(t) = e^{-\alpha*t}*u(t)$$
This signal is an energy signal.
$$E = 1/(2*\alpha)$$
- (c) $$x3(t) = e^{\alpha*t}*u(-t)$$
This is not a power signal because it is not periodic, and the power would be infinite.
This is not an energy signal because it never decays.


```python
import numpy as np
import matplotlib.pyplot as plt
# Question 2.9abc plots

t1 = np.linspace(-1, 1, 500)
x1 = 2 * np.cos(4 * np.pi * t1 + 2 * np.pi / 3)

plt.plot(t1, x1)
plt.title("x1(t) = 2cos(4œÄt + 2œÄ/3) (Power Signal)")
plt.xlabel("t")
plt.ylabel("x1(t)")
plt.grid(True)
plt.show()

# (b) x2(t) = e^(-Œ±t)u(t)
alpha = 1 # just has to be a positive constant
t2 = np.linspace(0, 5, 500) # Start at 0 due to u(t)
x2 = np.exp(-alpha * t2)

plt.plot(t2, x2)
plt.title("x2(t) = e^(-Œ±t)u(t) (Energy Signal)")
plt.xlabel("t")
plt.ylabel("x2(t)")
plt.grid(True)
plt.show()

# (c) x3(t) = e^(Œ±t)u(-t)
alpha = 1
t3 = np.linspace(-5, 0, 500)  # Negative t values due to u(-t)
x3 = np.exp(alpha * t3)

plt.plot(t3, x3)
plt.title("x3(t) = e^(Œ±t)u(-t) (Not Power or Energy Sig)")
plt.xlabel("t")
plt.ylabel("x3(t)")
plt.grid(True)
plt.show()
```
